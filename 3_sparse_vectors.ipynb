```python
from qdrant_client import QdrantClient, models
```

```python
# connect to qdrant cloud

from google.colab import userdata

client = QdrantClient(
    url = "https://b9d13b33-842f-4f65-8d2b-de07f146ebea.europe-west3-0.gcp.cloud.qdrant.io",
    api_key = userdata.get('qdrant_api_key')
)
```

```python
collection_name = "sparse_vectors_collection"

client.create_collection (
    collection_name = collection_name,
    sparse_vectors_config = {
        "sparse_vector": models.SparseVectorParams()
    }
)
```

### Insert sparse vectors into the collection

```python
collection_name = "sparse_vectors_collection"

client.upsert(
    collection_name = collection_name,
    points = [
        models.PointStruct(
            id = 1,
            payload = {},
            vector = {
                'sparse_vector': models.SparseVector(
                    indices = [1, 2, 3],
                    values = [0.2, -0.2, 0.2]

                )
            }
        ),
        models.PointStruct(
            id = 2,
            payload = {},
            vector = {
                'sparse_vector': models.SparseVector(
                    indices = [1, 5],
                    values = [0.1, 0.1]

                )
            }
        )
    ]
)
```

### Similarity Search on Sparse Vectors

```python
client.query_points(
    collection_name = collection_name,
    using = "sparse_vector",
    limit = 1,
    query = models.SparseVector(
        indices = [1, 3],
        values = [1, 1]
    ),
    with_vectors = True
)
```
